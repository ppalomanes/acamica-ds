```python
df['time_1'] = df['5K']
df['time_2'] = df['10K'] - df['5K']
df['time_3'] = df['15K'] - df['10K']
df['time_4'] = df['20K'] - df['15K']
df['time_5'] = df['25K'] - df['20K']
df['time_6'] = df['30K'] - df['25K']
df['time_7'] = df['35K'] - df['30K']
df['time_8'] = df['40K'] - df['35K']
df
```

```python
def toSeconds(cell):
    return cell.total_seconds()
def rowsToSecods(dataFrame, arr):
    for raw in arr:
        dataFrame[raw] = dataFrame.apply (lambda row: toSeconds(row[raw]),axis=1)
    return dataFrame
df.columns
```

```python
timeColumns = ['time_1', 'time_2',
       'time_3', 'time_4', 'time_5', 'time_6', 'time_7', 'time_8']
df = rowsToSecods(df, timeColumns)
df
```

```python
df.shape
```

```python
df = df[(df['time_1'] > 0) & 
   (df['time_2'] > 0) &
   (df['time_3'] > 0) &
   (df['time_4'] > 0) &
   (df['time_5'] > 0) &
   (df['time_6'] > 0) &
   (df['time_7'] > 0) &
   (df['time_8'] > 0)]
df.shape
```

```python
df = df[(df['time_1'] < df['time_1'].median()) & 
   (df['time_2'] < df['time_2'].median()) &
   (df['time_3'] < df['time_3'].median()) &
   (df['time_4'] < df['time_4'].median()) &
   (df['time_5'] < df['time_5'].median()) &
   (df['time_6'] < df['time_6'].median()) &
   (df['time_7'] < df['time_7'].median()) &
   (df['time_8'] < df['time_8'].median())]
df.shape
```

```python
df
```

```python
auxTable2 = df.groupby('Country').count().sort_values(by='Bib', ascending=False)
auxTable = auxTable2['Bib']
auxTable = auxTable.rename(columns = {'Bib':'niceRun'})
df1 = pd.concat([auxTable, auxTable1], axis=1, sort=True)
```

```python
df1[0] = df1[0].fillna(value=0)
df1
```

```python
def niceRunPorcentage(row):
    return int(round((row[0] * 100) / row[1]))

df1['niceRunPorcentage'] = df1.apply (lambda row: niceRunPorcentage(row),axis=1)
```

```python
df2 = df1.sort_values(by='Bib', ascending=False).sort_values(by='niceRunPorcentage', ascending=False)
df2
```
