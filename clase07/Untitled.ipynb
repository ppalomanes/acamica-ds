```python
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("movie_metadata.csv")
```

```python
df.T
```

```python
df.columns
```

```python
df = df[['director_name', 'actor_1_name', 'imdb_score']]
x1, y = df.shape
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(6)
```

```python
df = df.dropna()
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(6)
```

```python
x2, y = df.shape
print ("Original: {} \nLimpio: {} \nBorrados: {}".format(x1, x2, x1-x2))
```

```python
df_tuples = df.groupby(["director_name", "actor_1_name"])[["imdb_score"]].mean()
df_tuples.sort_values(by='imdb_score', ascending=False).head(20)
```

```python
df["participations"] = df["imdb_score"]
df_tuples2 = df.groupby(["director_name", "actor_1_name"]).agg({"participations" : "count", "imdb_score" : "mean"})
```

```python
type(df_tuples2)
```

```python
df_tuples2[df_tuples2["participations"] > 1].sort_values(by='imdb_score', ascending=False).head(20)
```
