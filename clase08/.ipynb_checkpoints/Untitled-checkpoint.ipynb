```python
df = pd.read_csv("movie_metadata.csv")
df.sample(10).T
```

## Las peliculas de mayor duracion son mejores

```python
df.dtypes
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(10,8)) 
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
           annot=True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.scatterplot(x='duration', y='num_user_for_reviews', data=df, ax=ax);
```

### Trajar con aquellos que tengan una buena base se seguidores

---
# Los mejores direcores (top 10) con al menos 5 peliculas dirigidas

```python
df["peliculas"] = df["imdb_score"]
df_d = df.groupby(["director_name"]).agg({"peliculas" : "count", "imdb_score" : "mean"})
```

```python
df_d[df_d["peliculas"] >= 5].sort_values(by='imdb_score', ascending=False).head(10)
```
