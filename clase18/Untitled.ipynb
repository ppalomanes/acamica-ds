# 1. Levantar Dataset

```python
import pandas as pd
import nltk
```

```python
df = pd.read_csv("dataset/tweets.csv")
```

```python
df
```

# 2. Seleccionar columna de texto(X) y label(y)

```python
df.columns
```

```python
df = df[['text', 'airline_sentiment']]
```

```python
df['airline_sentiment'].value_counts(normalize=True)
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

# 3. Label encoder de Y

```python
from sklearn.preprocessing import LabelEncoder
df_save_1 = df
```

```python
le = LabelEncoder() 
df['airline_sentiment']  = le.fit_transform(df['airline_sentiment'])
```

# 4. Dividir en train y test(Usar stratify para no desbalancer los datasets)

```python
X = df['text']
y = df['airline_sentiment']

# Realizá la separación a continuación en esta celda
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
X_train_save = X_train
X_test_save = X_test
```

# 5. Aplicar count vectorizer(BOW)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(analyzer='word', tokenizer=nltk.word_tokenize)
bow = vectorizer.fit_transform(X_train)
```

```python
bow
```

```python
vectorizer.get_feature_names()
```

```python
bow.toarray()
```

```python
X_train = bow
X_test = vectorizer.transform(X_test)
```

# 6. Entrenar con clasificador

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(n_estimators=200, n_jobs=-1)
rnd_clf.fit(X_train, y_train)
y_pred = rnd_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score,confusion_matrix,classification_report
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
f1_score(y_test, y_pred, average='micro')
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stops=set(stopwords.words('english'))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(analyzer='word', tokenizer=nltk.word_tokenize, stop_words=stops, ngram_range=(1,2))
tfidf_train = vectorizer.fit_transform(X_train_save)
tfidf_test =vectorizer.transform(X_test_save)
```

```python
rnd_clf = RandomForestClassifier(n_estimators=200, n_jobs=-1)
rnd_clf.fit(tfidf_train, y_train)
y_pred = rnd_clf.predict(tfidf_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
f1_score(y_test, y_pred, average='micro')
```
