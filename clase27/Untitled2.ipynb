```python
combi = train.append(test, ignore_index=True)
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt  
```

```python
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tweet'], "@[\w]*")
```

```python
combi['tidy_tweet'] = combi['tidy_tweet'].str.replace("[^a-zA-Z#]", " ")
```

```python
combi['tidy_tweet'] = combi['tidy_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
combi.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
stops = set(stopwords.words('english'))
```

```python
vectorizer = TfidfVectorizer(analyzer='word', 
                             tokenizer=nltk.word_tokenize, 
                             stop_words=stops, 
                             ngram_range=(1,3))
```

```python
vectorizer.fit(combi['tidy_tweet'])
```

```python
vectorizer.vocabulary_
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_ = combi[:31962]
test_ = combi[31962:]
```

```python
X = train_['tidy_tweet']
y = train_['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, stratify=y)
```

```python
X_train_v = vectorizer.transform(X_train)
X_test_v = vectorizer.transform(X_test)
```

```python
X_train_v.shape
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_clf = MLPClassifier(hidden_layer_sizes=(10,),verbose=True)
```

```python
mlp_clf.fit(X_train_v, y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score, classification_report
```

```python
y_pred = mlp_clf.predict(X_test_v)
```

```python
#confusion_matrix(y_test, y_pred)
```

```python
print("roc_auc score: {}".format(roc_auc_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
```

```python
X_all = vectorizer.transform(X)
y_all = y
```

```python
mlp_clf.fit(X_all, y_all)
```

```python
dtest = vectorizer.transform(test_['tidy_tweet'])
test_pred = mlp_clf.predict(dtest)
```

```python
test_['label'] = test_pred
```
